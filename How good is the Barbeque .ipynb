```python
import numpy as np
import pandas as pd
from io import StringIO
from sqlalchemy import create_engine
import pymysql
import matplotlib.pyplot as plt
%matplotlib inline
# import seaborn as sns
```

```python
engine = create_engine('mysql+pymysql://name:password@localhost/yelp_db')
```

```python
bus = pd.read_sql_table('business',engine,index_col='id')
```

```python
cat = pd.read_sql_table('category',engine,index_col= 'business_id')
```

```python
bbq = cat[cat.category == "Barbeque"]
bbq = bbq.join(bus,how = 'inner')
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(bbq.longitude,bbq.latitude,bbq.stars)
```
